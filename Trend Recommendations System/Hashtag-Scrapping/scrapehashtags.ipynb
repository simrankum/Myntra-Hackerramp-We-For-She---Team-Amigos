```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()


driver.get('https://www.hopperhq.com/blog/best-instagram-hashtags-fashion-bloggers/')  
# similarly we can use more websites url to extract more hashtags 


h2_elements = driver.find_elements(By.TAG_NAME, 'h2')

spans = []


for h2_element in h2_elements: 

    p_elements = h2_element.find_elements(By.XPATH, './following-sibling::p')
    
    for p_element in p_elements:
        span_elements = p_element.find_elements(By.TAG_NAME, 'span')
        
        for span in span_elements: 

            spans.append(span.text)

combined_text = " ".join(spans)
hashtags_list = combined_text.split()

print("Hashtags extracted:")
for hashtag in hashtags_list:
    print(hashtag)

# saving the hashtags in a file for further processing
with open('hashtags_list.txt', 'w') as file:
    for hashtag in hashtags_list:
        file.write(f"{hashtag}\n")


driver.quit()
```
